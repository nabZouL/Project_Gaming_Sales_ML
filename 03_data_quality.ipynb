# Clean de MC

```python
mc = pd.read_csv('MC_final.csv')
mc = mc.drop('Date', axis = 1)
```

```python
mc.info()
```

```python
# MODIFICATION NOMS DES COLONNES

mc = mc.rename(columns = {'Note MC':'Test_MC','Note Players MC':'Players_MC'})

# REGROUPEMENT PAR PLATEFORMES

platform_replace = {'PlayStation 4' : 'Playstation', 'Xbox 360' : 'Xbox','PlayStation 2' : 'Playstation',
                    'PlayStation 3' : 'Playstation', 'PlayStation 5' : 'Playstation','Xbox One' : 'Xbox',
                    'PSP':'PlayStation','Game Boy Advance':'GB','PlayStation Vita' : 'Playstation','3DS':'DS',
                    'GameCube':'Autre','Xbox Series X':'Xbox','Wii U':'Wii', 'Dreamcast':'Autre',
                    'Nintendo 64':'Nintendo','Stadia':'Autre'}

mc['Platform'] = mc['Platform'].replace(platform_replace)

# HOMOGENEISATION DE LA CASSE

mc['Name'] = mc['Name'].apply(lambda x: x.upper())
mc['Platform'] = mc['Platform'].apply(lambda x: x.upper())

# RETRAIT DES ELEMENTS DE TYPE (a-zA-Z0-9) DANS LES TITRES DE JEUX

for i, element in enumerate(mc['Name']):
  if '(' in element :
    r1 = re.findall(r"\(.*?\)", element)
    mc['Name'][i] = element.replace(r1[0], '').strip()

# GROUPBY NAME et PLATFORM : PS1 = Playstation ; PS4 = Playstation pour un même jeu

mc = mc.groupby(['Name','Platform']).mean().reset_index()
mc.head()
```

# Clean de JV

```python
jv = pd.read_csv('dataframe_jeuxvideos.csv')
jv = jv.drop('Unnamed: 0', axis = 1)
display(jv.head())
print(jv.shape)
```

```python
# MODIFICATION NOMS DES COLONNES
jv = jv.rename(columns = {'titres' : 'Name', 
                          'test':'Test_JV', 
                          'avis':'Players_JV'})

# HOMOGENEISATION DE LA CASSE
jv['Name'] = jv['Name'].apply(lambda x : x.upper())

# RETRAIT DES ELEMENTS DE TYPE (a-zA-Z0-9) DANS LES TITRES DE JEUX
for i, element in enumerate(jv['Name']):
  if '(' in element :
    r1 = re.findall(r"\(.*?\)", element)
    jv['Name'][i] = element.replace(r1[0], '').strip()

# MODIFICATION DES VALEURS DES NOTES

to_replace = {'- -/20' : '0/20', 'test non renseigné' : '0/20',             # remplacement des notes non renseignées
              'avis non renseigné' : '0/20','Non renseigné':'0/20',         # par des 0 (elles deviendronts des NaN à terme)
              '18,5/20':'18.5/20'}                                           

jv['Test_JV'] = jv['Test_JV'].replace(to_replace)
jv['Players_JV'] = jv['Players_JV'].replace(to_replace)

jv['Test_JV'] = jv['Test_JV'].apply(lambda x : x[:-3]).astype(float)        # on tronque la partie '/20' des notes : 18/20 => 18
jv['Players_JV'] = jv['Players_JV'].apply(lambda x : x[:-3]).astype(float)

to_replace_nan = {0 : np.nan}                                               # On remplace les 0 par des NaN pour rester fidèle à la réalité
jv['Test_JV'] = jv['Test_JV'].replace(to_replace_nan)
jv['Players_JV'] = jv['Players_JV'].replace(to_replace_nan)

# HOMOGENEISATION DES TITRES : POSITION DU CARACTERE ':'

for i, element in enumerate(jv['Name']) :
  if ' : ' in element :
    r2 = re.findall(r"\ :.*?\ ", element)
    jv['Name'][i] = element.replace(r2[0], ': ')

# RECUPERATION DES PLATEFORMES DANS LES TITRES DES JEUX

plateforme_jv = []
for element in jv['Name']:
  if ' SUR ' in element :
    plateforme_jv.append(element.split(' SUR ')[1])
  else :
    plateforme_jv.append(np.nan)

jv['Platform'] = plateforme_jv

plateforme_replace = {'PS1':'PLAYSTATION','PS2':'PLAYSTATION','PS3':'PLAYSTATION','PS4':'PLAYSTATION',
                      'PS5':'PLAYSTATION','PSP':'PLAYSTATION','VITA':'PLAYSTATION','XBOX SERIES':'XBOX',
                      '3DS':'DS','360':'XBOX','WIIU':'WII','GBA':'GB','IOS':'MOBILE','ONE':'XBOX','NGC':'GB',
                      'N64':'NINTENDO','NES':'NINTENDO','SNES':'NINTENDO','WEB':'MOBILE','GBC':'GB','INTERNET':'MOBILE',
                      'DCAST':'AUTRE','MD':'AUTRE','SATURN':'AUTRE','NGAGE':'AUTRE','MS':'AUTRE','LINUX':'MOBILE','NEW 3DS':'DS'}


# MODIFICATION DES TITRES

jv['Name'] = jv['Name'].apply(lambda x : x.split(' SUR ')[0])

jv.head()
```

```python
for cols in ['Name','Test_JV','Players_JV','Platform']:
  print('Nombre de NaN dans',cols,' : ',jv[cols].isna().sum(), 
        '\n', '% de NaN dans',cols, ' : ',(100 * jv[cols].isna().sum() / len(jv[cols])).round(2),'%')
```

```python
jv = jv.drop('Platform', axis = 1)
```

```python
vgc = pd.read_csv('vg_chartz_cleaned.csv')
vgc.head()
```

```python
vgc.info()
```

```python
vgc = vgc.rename(columns = {'Global Sales':'Global_Sales','Date de sortie':'Year'})

for i, element in enumerate(vgc['Name']) :
  if '(' in element :
    r2 = re.findall(r"\(.*?\)", element)
    vgc['Name'][i] = element.replace(r2[0], '').strip()

for i, element in enumerate(vgc['Name']) :
  if 'Read the review' in element :
    r3 = re.findall("Read the review", element)
    vgc['Name'][i] = element.replace(r3[0], '').strip()

vgc = vgc.groupby(['Name','Platform','Publisher']).agg({'Global_Sales':'sum','Year':'mean'}).sort_values(by = 'Global_Sales', ascending = False).reset_index()

vgc['Global_Sales'] = vgc['Global_Sales'].replace({0.0:np.nan})
vgc = vgc.dropna()

for col in ['Name','Platform','Publisher']:
  vgc[col] = vgc[col].apply(lambda x : x.upper())

vgc.head()
```
